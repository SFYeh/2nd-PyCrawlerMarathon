1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
import json
data = json.loads(response)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print(d['title'])
    print('貼文時間',d['createdAt'])
    print('留言人數',d['commentCount'])
    print('按讚',d['likeCount'])  
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
populer_comm=0
popular_like=0
poor_comm=0
poor_like=0
```

```python
#r = requests.get('https://www.dcard.tw/_api/forums/pet/posts? popular=true')  <---熱門與否標示在網址上！！！
for d in data:
    populer_comm=(populer_comm+d['commentCount'])  #populer_comm+=d['commentCount']) 
    popular_like=(popular_like+d['likeCount'])
print(populer_comm/len(data))
print(popular_like/len(data))
```

```python
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
data = json.loads(response)
for d in data:
    poor_comm=(poor_comm+d['commentCount'])
    poor_like=(poor_like+d['likeCount'])
print(poor_comm/len(data))
print(poor_like/len(data))
```
